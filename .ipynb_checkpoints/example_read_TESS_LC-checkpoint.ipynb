## Example of reading in TESS light curve data based on simulated data

```python
from imports import *
%matplotlib inline
```

```python
# get data
hdu = fits.open('ETE-6/tid/00/000/057/tess2019128220341-0000000005712108-0016-s_lc.fits')
hdu[0].header
```

```python
# get time-series
print hdu[1].columns
t, f, ef = hdu[1].data['TIME'], hdu[1].data['SAP_FLUX'], hdu[1].data['SAP_FLUX_ERR']
f2, ef2 = hdu[1].data['PDCSAP_FLUX'], hdu[1].data['PDCSAP_FLUX_ERR']
```

```python
plt.figure(figsize=(7,8))
norm, norm2 = np.median(f[np.isfinite(f)]), np.median(f2[np.isfinite(f2)])
plt.errorbar(t, f/norm, ef/norm, fmt='k.'), plt.axhline(1)
plt.errorbar(t, f2/norm2-.02, ef2/norm2, fmt='b.'), plt.axhline(1-.02)
plt.xlabel('BJD - 2457000, days'), plt.ylabel('Normalized flux')
```
